```python
import pandas as pd
from sqlalchemy import create_engine

# Read csv
df = pd.read_csv("C:/Users/ASTA/OneDrive/Desktop/POWERBI NOTES/project datas saving/dairy product data set.csv")

#  Connect to MySQL 
engine = create_engine("mysql+pymysql://root:sharmila123@localhost:3306/project")

# Step 3: Upload to SQL
df.to_sql("dairy_product", con=engine, if_exists="replace", index=False)
```

```python
query = "SELECT * FROM dairy_product LIMIT 5"
result = pd.read_sql(query, con=engine)

print(result)
```

```python
query = "SELECT COUNT(*) AS total_records FROM dairy_product"
result = pd.read_sql(query, con=engine)

print(result)
```

```python
query = "DESCRIBE dairy_product"
result = pd.read_sql(query, con=engine)

print(result)
```

```python
query = """
SELECT 
    `Location`,
    SUM(`Approx. Total Revenue(INR)`) AS total_revenue
FROM dairy_product
GROUP BY `Location`
ORDER BY total_revenue DESC
"""
result = pd.read_sql(query, con=engine)

print(result)
```

```python
query = """
SELECT 
    `Product Name`,
    SUM(`Quantity Sold (liters/kg)`) AS total_quantity
FROM dairy_product
GROUP BY `Product Name`
ORDER BY total_quantity DESC
LIMIT 5
"""
result = pd.read_sql(query, con=engine)

print(result)
```

```python
query = """
SELECT 
    `Brand`,
    SUM(`Quantity Sold (liters/kg)`) AS total_sales,
    SUM(`Approx. Total Revenue(INR)`) AS total_revenue
FROM dairy_product
GROUP BY `Brand`
ORDER BY total_revenue DESC
"""
result = pd.read_sql(query, con=engine)

print(result)
```

```python
query = """
SELECT 
    `Storage Condition`,
    AVG(`Shelf Life (days)`) AS avg_shelf_life,
    SUM(`Quantity Sold (liters/kg)`) AS total_sales
FROM dairy_product
GROUP BY `Storage Condition`
"""
result = pd.read_sql(query, con=engine)

print(result)
```

```python
query = """
SELECT 
    `Customer Location`,
    SUM(`Approx. Total Revenue(INR)`) AS total_revenue
FROM dairy_product
GROUP BY `Customer Location`
ORDER BY total_revenue DESC
"""
result = pd.read_sql(query, con=engine)

print(result)
```

```python
query = """
SELECT 
    `Product Name`,
    COUNT(*) AS near_expiry_count
FROM dairy_product
WHERE `Shelf Life (days)` < 5
GROUP BY `Product Name`
ORDER BY near_expiry_count DESC
"""
result = pd.read_sql(query, con=engine)

print(result)
```

```python
query = """
SELECT 
    `Location`,
    SUM(`Quantity Sold (liters/kg)`) AS demand
FROM dairy_product
GROUP BY `Location`
ORDER BY demand DESC
"""
result = pd.read_sql(query, con=engine)

print(result)
```

```python
query = """
SELECT 
    `Product Name`,
    `Quantity in Stock (liters/kg)`,
    `Minimum Stock Threshold (liters/kg)`
FROM dairy_product
WHERE `Quantity in Stock (liters/kg)` < `Minimum Stock Threshold (liters/kg)`
"""
result = pd.read_sql(query, con=engine)

print(result)
```

```python
query = """
SELECT 
    `Sales Channel`,
    SUM(`Approx. Total Revenue(INR)`) AS total_revenue
FROM dairy_product
GROUP BY `Sales Channel`
ORDER BY total_revenue DESC
"""
result = pd.read_sql(query, con=engine)

print(result)
```

```python
query = """
SELECT 
    SUBSTRING(`Date`, 1, 7) AS month,
    SUM(`Approx. Total Revenue(INR)`) AS monthly_revenue
FROM dairy_product
GROUP BY month
ORDER BY monthly_revenue DESC
LIMIT 5
"""
result = pd.read_sql(query, con=engine)

print(result)
```

```python
query = """
SELECT 
    SUBSTRING(`Date`, 1, 4) AS year,
    SUM(`Approx. Total Revenue(INR)`) AS yearly_revenue
FROM dairy_product
GROUP BY year
ORDER BY yearly_revenue DESC
LIMIT 5
"""
result = pd.read_sql(query, con=engine)

print(result)
```
